```python
def calculate_health(temp, vibration, hours):
    score = 100

    if temp > 80:
        score -= 30
    elif temp > 60:
        score -= 15

    if vibration > 5:
        score -= 30
    elif vibration > 3:
        score -= 15

    if hours > 1000:
        score -= 25
    elif hours > 500:
        score -= 10

    return score


def machine_status(score):
    if score >= 70:
        return "HEALTHY"
    elif score >= 40:
        return "WARNING - Schedule Maintenance"
    else:
        return "HIGH FAILURE RISK - Immediate Action Required"


print("=== Predictive Maintenance System ===")

machine = input("Machine Name: ")
temp = float(input("Temperature (°C): "))
vibration = float(input("Vibration Level (mm/s): "))
hours = int(input("Running Hours: "))

score = calculate_health(temp, vibration, hours)
status = machine_status(score)

print("\n------ MACHINE REPORT ------")
print(f"Machine: {machine}")
print(f"Health Score: {score}/100")
print(f"Status: {status}")
```
